This notebook contains experiment where number of features when splitting (max_features) is varying, everything else is fixed

```python
import numpy as np 
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../data/spambase_augmented.csv')
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
print('df shape : ' , df.shape)
print('X shape : ' , X.shape)
print('y shape : ', y.shape)
results = {}
```

max_features = log2

```python
train_error = []
test_error = []
train_sizes = np.linspace(0.1, 0.9, 9)

for train_size in train_sizes:
    X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=train_size,random_state=42)
    clf = RandomForestClassifier(max_features='log2',n_jobs=-1)
    clf.fit(X_train,y_train)

    #train error
    y_train_pred = clf.predict(X_train)
    cur_training_err = 1 - accuracy_score(y_train, y_train_pred)

    # Compute test error
    y_test_pred = clf.predict(X_test)
    cur_test_error = 1 - accuracy_score(y_test, y_test_pred)

    train_error.append(cur_training_err)
    test_error.append(cur_test_error)

plt.figure(figsize=(8,6))
plt.plot(train_sizes, train_error, label="Training Error", marker='o', linestyle='-')
plt.plot(train_sizes, test_error, label="Test Error ", marker='s', linestyle='--')
plt.xlabel("Training Set Size")
plt.ylabel("Error Rate")    
plt.title(f"max_features = log2")
plt.legend()
plt.grid(True)
plt.show()
results['log2'] = [train_error,test_error]
```

```python
print(results)
```
